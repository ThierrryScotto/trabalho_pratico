```python
import pandas as pd
import glob
import os

# Defina o caminho da pasta onde os datasets estão localizados
caminho_pasta = 'previsão\POA_padronizados\*.csv'  # Altere para o caminho correto

# Use glob para obter todos os arquivos CSV no caminho especificado
arquivos = glob.glob(caminho_pasta)

# Lista para armazenar os DataFrames individuais
dfs = []

# Loop através de cada arquivo encontrado
for arquivo in arquivos:
    # Lê o dataset e define "Data" como índice
    df = pd.read_csv(arquivo, sep=';', index_col='DATA', parse_dates=['DATA'])
    
    # Adiciona o DataFrame à lista
    dfs.append(df)

# Une todos os DataFrames com base no índice
df_unificado = pd.concat(dfs)

# Ordena pelo índice de data, caso os dados não estejam em ordem
df_unificado.sort_index(inplace=True)

# Resetar o índice se não quiser que "Data" seja o índice
# df_unificado.reset_index(inplace=True)

# Salvar o DataFrame unificado em um novo arquivo CSV
df_unificado.to_csv('INMET.csv', sep=';')

print("Datasets combinados com sucesso em 'INMET.csv'")
```
